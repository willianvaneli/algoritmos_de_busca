```
import pandas as pd
import matplotlib.pyplot as plt

# Defining CSV headers
headers = ['algorithm','number of elements','time']

# Opening CSV from github
selectionsort = pd.read_csv("https://raw.githubusercontent.com/v1eira/tpa-trab2/master/output/selectionsort.csv", names=headers)
```

```
# Creating list with the sizes of the CSVs analysed by the algorithm and sorting it

sizes = [
10,
100,
1000,
10000,
100000,
1000000,
25,
250,
2500,
25000,
250000,
2500000,
50,
500,
5000,
50000,
500000,
5000000,
75,
750,
7500,
75000,
750000,
7500000]

sizes.sort()
```

```
# Create Selection Sort DataFrame with min, mean and max time values

ss_times = {
    "number of elements":[],
    "min time":[],
    "avg time":[],
    "max time":[]
}

for i in sizes:
    min_time = selectionsort.loc[(selectionsort['number of elements']) == i,"time"].min()
    mean_time = selectionsort.loc[(selectionsort['number of elements']) == i,"time"].mean()
    max_time = selectionsort.loc[(selectionsort['number of elements']) == i,"time"].max()
    ss_times["number of elements"].append(i)
    ss_times["min time"].append(min_time)
    ss_times["avg time"].append(mean_time)
    ss_times["max time"].append(max_time)

df = pd.DataFrame({
    "number of elements": ss_times["number of elements"],
    "minimum time": ss_times["min time"],
    "average time": ss_times["avg time"],
    "maximum time": ss_times["max time"]
})
```

```
# Plot graphs

ax = plt.gca()

df.loc[(df['number of elements']) <= 50000].plot(kind='line',x='number of elements',y='minimum time', ax=ax)
df.loc[(df['number of elements']) <= 50000].plot(kind='line',x='number of elements',y='average time', color='red', ax=ax)
df.loc[(df['number of elements']) <= 50000].plot(kind='line',x='number of elements',y='maximum time', color='green', ax=ax)

plt.show()
```
